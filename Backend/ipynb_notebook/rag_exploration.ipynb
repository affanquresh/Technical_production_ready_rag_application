```python
import pandas as pd
import re

df = pd.read_csv("/Users/affanqureshi/Desktop/affan/Production_ready_RAG_application/synthetic_knowledge_items.csv")

def clean_text(text):
    # Remove markdown symbols
    text = re.sub(r"\*\*", "", text)   # remove **
    text = re.sub(r"\*", "", text)     # remove *
    
    # Replace line breaks with space
    text = text.replace("\n", " ")
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text)
    
    return text.strip()

df["clean_ki_text"] = df["ki_text"].apply(clean_text)
```

```python
import re
import pandas as pd

df = pd.read_csv("/Users/affanqureshi/Desktop/affan/Production_ready_RAG_application/synthetic_knowledge_items.csv")
```

```python
def chunk_text(text, doc_id, topic):
    chunks = []

    # Normalize text
    text = text.strip()

    # Split by sections using keywords
    section_patterns = [
        r"\*\*Prerequisites\*\*",
        r"\*\*Step \d+.*?\*\*",
        r"\*\*Troubleshooting.*?\*\*",
        r"\*\*Tips.*?\*\*"
    ]

    # Combine patterns
    combined_pattern = "|".join(section_patterns)

    # Split text while keeping delimiters
    splits = re.split(f"({combined_pattern})", text)

    current_section = "general"
    step_number = None

    for part in splits:
        part = part.strip()
        if not part:
            continue

        # Detect section type
        if "Prerequisites" in part:
            current_section = "prerequisites"
            step_number = None
            continue

        elif "Step" in part:
            current_section = "step"
            match = re.search(r"Step (\d+)", part)
            step_number = int(match.group(1)) if match else None
            continue

        elif "Troubleshooting" in part:
            current_section = "troubleshooting"
            step_number = None
            continue

        elif "Tips" in part:
            current_section = "tips"
            step_number = None
            continue

        # Otherwise it's actual content
        chunk = {
            "content": part,
            "metadata": {
                "doc_id": str(doc_id),
                "topic": topic,
                "section": current_section,
                "step_number": step_number
            }
        }

        chunks.append(chunk)

    # Fallback if nothing split properly
    if len(chunks) == 0:
        paragraphs = text.split("\n\n")
        for para in paragraphs:
            para = para.strip()
            if para:
                chunks.append({
                    "content": para,
                    "metadata": {
                        "doc_id": str(doc_id),
                        "topic": topic,
                        "section": "general",
                        "step_number": None
                    }
                })

    return chunks
```

## Build RAG pipeline with Functions

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")

llm_model = ChatGroq(
    model="openai/gpt-oss-120b",
    api_key=groq_api_key,
    temperature=0
)
```

```python
def retrieve_chunks(query, model, embeddings, texts, metadatas, top_k=3):
    
    query_embedding = model.encode([query])[0]
    scores = cosine_similarity([query_embedding], embeddings)[0]
    
    top_indices = np.argsort(scores)[-top_k:][::-1]
    
    retrieved = []
    for idx in top_indices:
        retrieved.append({
            "content": texts[idx],
            "metadata": metadatas[idx],
            "score": scores[idx]
        })
    
    return retrieved
```

```python
def build_context(retrieved_chunks):
    context = ""
    
    for i, chunk in enumerate(retrieved_chunks):
        context += f"[Chunk {i+1}]\n{chunk['content']}\n\n"
    
    return context
```

```python
def generate_answer(query, retrieved_chunks):
    
    context = build_context(retrieved_chunks)

    prompt = f"""
Answer the question using ONLY the context below.
If the answer is not present, say "I don't know".

Context:
{context}

Question:
{query}

Answer:
"""

    response = llm_model.invoke(prompt)
    return response.content
```

```python
def show_sources(retrieved_chunks):
    print("\nSources:\n")
    
    for i, chunk in enumerate(retrieved_chunks):
        preview = " ".join(chunk["content"].split("\n")[:3])
        
        print(f"[Source {i+1}]")
        print("Score:", chunk["score"])
        print("Preview:", preview)
        print("Metadata:", chunk["metadata"])
        print("-" * 50)
```

```python
def run_rag_pipeline(query, top_k=3):
    
    retrieved_chunks = retrieve_chunks(
        query=query,
        model=model,          # your embedding model
        embeddings=embeddings,
        texts=texts,
        metadatas=metadatas,
        top_k=top_k
    )
    
    answer = generate_answer(query, retrieved_chunks)
    
    print("\nFinal Answer:\n", answer)
    show_sources(retrieved_chunks)
```

```python
run_rag_pipeline("what are the requirements the remote workers need to meet for VPN access?")
```

## lets make a hybrid Retriever that works with bm25 and embedding both

```python
load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")

llm_model = ChatGroq(
    model="openai/gpt-oss-120b",
    api_key=groq_api_key,
    temperature=0
)
```

```python
def hybrid_retrieve(query, model, embeddings, texts, metadatas, bm25, top_k, alpha=0.5):
    
    # -------- 1. Embedding Scores --------
    query_embedding = model.encode([query])[0]
    emb_scores = cosine_similarity([query_embedding], embeddings)[0]
    
    # Normalize embedding scores (0 → 1)
    emb_scores = (emb_scores - emb_scores.min()) / (emb_scores.max() - emb_scores.min() + 1e-8)
    
    
    # -------- 2. BM25 Scores --------
    tokenized_query = query.lower().split()
    bm25_scores = bm25.get_scores(tokenized_query)
    
    # Normalize BM25 scores (0 → 1)
    bm25_scores = (bm25_scores - bm25_scores.min()) / (bm25_scores.max() - bm25_scores.min() + 1e-8)
    
    
    # -------- 3. Combine Scores --------
    final_scores = alpha * emb_scores + (1 - alpha) * bm25_scores
    
    
    # -------- 4. Get Top Results --------
    top_indices = np.argsort(final_scores)[-top_k:][::-1]
    
    
    # -------- 5. Build Output --------
    retrieved_chunks = []
    
    for idx in top_indices:
        retrieved_chunks.append({
            "content": texts[idx],
            "metadata": metadatas[idx],
            "score": final_scores[idx]
        })
    
    return retrieved_chunks
```

```python
def build_context(retrieved_chunks):
    context = ""
    
    for i, chunk in enumerate(retrieved_chunks):
        context += f"[Chunk {i+1}]\n{chunk['content']}\n\n"
    
    return context
```

```python
def generate_answer(query, retrieved_chunks):
    
    context = build_context(retrieved_chunks)

    prompt = f"""
Answer the question using ONLY the context below.
If the answer is not present, say "I don't know".

Context:
{context}

Question:
{query}

Answer:
"""

    response = llm_model.invoke(prompt)
    return response.content
```

```python
def show_sources(retrieved_chunks):
    print("\nSources:\n")
    
    for i, chunk in enumerate(retrieved_chunks):
        preview = " ".join(chunk["content"].split("\n")[:3])
        
        print(f"[Source {i+1}]")
        print("Score:", chunk["score"])
        print("Preview:", preview)
        print("Metadata:", chunk["metadata"])
        print("-" * 50)
```

```python
def run__hybrid_rag_pipeline(query, top_k=10):
    
    retrieved_chunks = hybrid_retrieve(
        query=query,
        model=model,          # your embedding model
        embeddings=embeddings,
        texts=texts,
        metadatas=metadatas,
        bm25=bm25,
        top_k=top_k
    )
    
    answer = generate_answer(query, retrieved_chunks)
    
    print("\nFinal Answer:\n", answer)
    show_sources(retrieved_chunks
    )
```

```python
def rerank_chunks(query, retrieved_chunks, top_k):
    
    # Step 1: create (query, chunk) pairs
    pairs = [(query, chunk["content"]) for chunk in retrieved_chunks]
    
    # Step 2: get relevance scores
    scores = re_ranker.predict(pairs)
    
    # Step 3: attach scores to chunks
    for i, chunk in enumerate(retrieved_chunks):
        chunk["rerank_score"] = float(scores[i])
    
    # Step 4: sort by rerank score (descending)
    reranked = sorted(
        retrieved_chunks,
        key=lambda x: x["rerank_score"],
        reverse=True
    )
    
    # Step 5: return top_k
    return reranked[:top_k]
```

```python
def show_reranked_sources(retrieved_chunks):
    print("\nSources:\n")
    
    for i, chunk in enumerate(retrieved_chunks):
        preview = " ".join(chunk["content"].split("\n")[:3])
        
        print(f"[Source {i+1}]")
        print("Score:", chunk["score"])
        print("Preview:", preview)
        print("Metadata:", chunk["metadata"])
        print("Re-rank Score:", chunk.get("rerank_score", "N/A"))
        print("-" * 50)
```

```python
def run_reranked_rag_pipeline(query, top_k=5):
    
    # Step 1: Hybrid retrieval
    retrieved_chunks = hybrid_retrieve(
        query=query,
        model=model,
        embeddings=embeddings,
        texts=texts,
        metadatas=metadatas,
        bm25=bm25,
        top_k=10
    )
    
    # Step 2: Re-ranking
    final_chunks = rerank_chunks(query, retrieved_chunks, top_k=top_k)
    
    # Step 3: Generate answer
    answer = generate_answer(query, final_chunks)
    
    print("\nFinal Answer:\n", answer)
    show_reranked_sources(final_chunks)
```

```python
def keyword_match_score(answer, keywords):
    answer = answer.lower()
    
    match_count = 0
    
    for keyword in keywords:
        # split keyword into words
        words = keyword.split()
        
        # check if ALL words exist somewhere
        if all(word in answer for word in words):
            match_count += 1
    
    return match_count / len(keywords)
```

```python
def evaluate_rag(qa_data, use_rerank=True):
    
    scores = []
    
    for item in qa_data:
        query = item["question"]
        keywords = item["keywords"]
        
        # Step 1: retrieve
        retrieved_chunks = hybrid_retrieve(
            query=query,
            model=model,
            embeddings=embeddings,
            texts=texts,
            metadatas=metadatas,
            bm25=bm25,
            top_k=10
        )
        
        # Step 2: rerank (optional)
        if use_rerank:
            final_chunks = rerank_chunks(query, retrieved_chunks, top_k=3)
        else:
            final_chunks = retrieved_chunks[:3]
        
        # Step 3: generate answer
        answer = generate_answer(query, final_chunks)
        
        # Step 4: score
        score = keyword_match_score(answer, keywords)
        scores.append(score)
        
        print("Question:", query)
        print("Score:", score)
        print("-"*40)
    
    return sum(scores) / len(scores)
```

```python
def evaluate_correctness(question, reference, answer):
    
    prompt = f"""
You are evaluating an AI system.

QUESTION: {question}

GROUND TRUTH:
{reference}

MODEL ANSWER:
{answer}

Is the model answer factually correct?
Answer only YES or NO.
"""
    
    result = llm_model.invoke(prompt).content
    
    return "yes" in result.lower()
```

```python
def evaluate_groundedness(answer, documents):
    
    context = "\n\n".join([doc["content"] for doc in documents])
    
    prompt = f"""
FACTS:
{context}

ANSWER:
{answer}

Is the answer fully supported by the facts?
Answer only YES or NO.
"""
    
    result = llm_model.invoke(prompt).content
    
    return "yes" in result.lower()
```

```python
def evaluate_system(qa_data):
    
    correct_scores = []
    grounded_scores = []
    
    for item in qa_data:
        
        query = item["question"]
        reference = item["reference"]
        
        result = run_reranked_rag_pipeline_with_docs(query)
        
        answer = result["answer"]
        docs = result["documents"]
        
        correct = evaluate_correctness(query, reference, answer)
        grounded = evaluate_groundedness(answer, docs)
        
        correct_scores.append(int(correct))
        grounded_scores.append(int(grounded))
        
        print(f"Q: {query}")
        print(f"Correct: {correct}, Grounded: {grounded}")
        print("-"*50)
    
    print("\nFinal Scores:")
    print("Correctness:", sum(correct_scores)/len(correct_scores))
    print("Groundedness:", sum(grounded_scores)/len(grounded_scores))
```

```python
from langchain_community.vectorstores import Chroma

vector_store = Chroma(
    persist_directory="new_chroma_db",
    embedding_function=embedding_model
)
```

```python
import numpy as np

def normalize(arr):
    return (arr - arr.min()) / (arr.max() - arr.min() + 1e-8)


def hybrid_retrieve(query, bm25, vector_store, texts, metadatas, top_k=5, alpha=0.5):
    
    formatted_query = "Represent this sentence for searching relevant passages: " + query
    
    # BM25
    bm25_scores = bm25.get_scores(query.split())
    
    # Chroma
    docs = vector_store.similarity_search_with_score(formatted_query, k=len(texts))
    
    chroma_scores = np.zeros(len(texts))
    
    for doc, score in docs:
        idx = doc.metadata["id"]
        chroma_scores[idx] = score
    
    # Convert distance → similarity
    chroma_scores = 1 / (1 + chroma_scores)
    
    # Normalize
    bm25_norm = normalize(bm25_scores)
    chroma_norm = normalize(chroma_scores)
    
    # Hybrid
    final_scores = alpha * chroma_norm + (1 - alpha) * bm25_norm
    
    # Top K
    top_indices = np.argsort(final_scores)[-top_k:][::-1]
    
    results = []
    
    for idx in top_indices:
        results.append({
            "content": texts[idx],
            "metadata": metadatas[idx],
            "bm25_score": bm25_scores[idx],
            "chroma_score": chroma_scores[idx],
            "final_score": final_scores[idx]
        })
    
    return results
```

```python
def build_context(chunks):
    context = ""
    
    for i, chunk in enumerate(chunks):
        context += f"[Chunk {i+1}]\n{chunk['content']}\n\n"
    
    return context
```

```python
from langchain_groq import ChatGroq
import os

llm = ChatGroq(
    model="openai/gpt-oss-120b",
    api_key=os.getenv("GROQ_API_KEY")
)

def generate_answer(query, chunks):
    
    context = build_context(chunks)
    
    prompt = f"""
You are an AI assistant.

Answer ONLY from the context below.
If the answer is not present, say "I don't know".

Context:
{context}

Question:
{query}

Answer:
"""
    
    response = llm.invoke(prompt)
    
    return response.content
```
